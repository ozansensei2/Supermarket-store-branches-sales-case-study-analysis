#### lets answer some basic questions

```python
#which store has biggest and lowest area
biggest_area = df[(df['Store_Area'] == df['Store_Area'].max())]
lowest_area = df[(df['Store_Area'] == df['Store_Area'].min())]

print (biggest_area)
print (lowest_area)
```

```python
#which store has most and least Items_Available
most_items = df[(df['Items_Available'] == df['Items_Available'].max())]
least_item = df[(df['Items_Available'] == df['Items_Available'].min())]

print (most_items)
print (least_item)
```

```python
#which store has most and least Daily_Customer_Count
most_customer = df[(df['Daily_Customer_Count'] == df['Daily_Customer_Count'].max())]
least_customer = df[(df['Daily_Customer_Count'] == df['Daily_Customer_Count'].min())]

print (most_customer)
print (least_customer)
```

```python
#which store has most and least Store_Sales
most_sales = df[(df['Store_Sales'] == df['Store_Sales'].max())]
least_sales = df[(df['Store_Sales'] == df['Store_Sales'].min())]

print (most_sales)
print (least_sales)
```

#####  lets check on scatter plot with reggesion line

```python
##
sns.lmplot(x ="Store_Area" ,y= 'Items_Available', data = df, lowess = True, aspect = 3, line_kws = {'color':"red"}).set(title="Store Area vs Items")
plt.show()
```

```python
sns.lmplot(x ="Store_Sales" ,y= 'Items_Available', data = df, lowess = True, aspect = 3, line_kws = {'color':"red"}).set(title="Store Sales vs Items")
plt.show()
```

```python
sns.lmplot(x ="Store_Sales" ,y= 'Daily_Customer_Count', data = df, lowess = True, aspect = 3, line_kws = {'color':"red"}).set(title="Store Sales vs Items")
plt.show()
```

```python
sns.lmplot(x ="Store_Sales" ,y= 'Store_Area', data = df, lowess = True, aspect = 3, line_kws = {'color':"red"}).set(title="Store Sales vs Items")
plt.show()
```

#1 The Regression line shows that store area and items highly correlated it expected.

#2 The Regression line shows sales is more when there are more items

#3 The Regression line almost plain,sales and customer count arent depentent each other

#4 The Regression line describes that sales are a little bit more when store area is high
